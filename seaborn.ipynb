```python
k=df.dropna() #remove null value records temporarily
k.shape # no.of rows and cols after removing null value records
```

```python
df.drop('deck', axis=1, inplace=True) # to drop 'deck' col from dataset as it is >20 of loss
df.columns # check the col is dropped or not

```

```python
df['age'] = df['age'].fillna(df['age'].median())
df['age']
```

```python
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])# fill the missing values in embarked with mode
df['embarked']
```

```python
df['embark_town']= df['embark_town'].fillna(df['embark_town'].mode()[0]) # fill the missing values in embark_town with mode
df['embark_town']
```
